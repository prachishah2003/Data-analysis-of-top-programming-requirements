# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
QUERY = """
SELECT ConvertedComp
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
plt.hist(df, bins=100, color='blue', alpha=0.7, edgecolor='black')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
plt.boxplot(df.dropna())
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import seaborn as sns

QUERY = """
SELECT Age,WorkWeekHrs
FROM master
WHERE WorkWeekHrs<=75
"""

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)

plt.title('Scatter Plot of Age vs WorkweekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
WHERE WorkWeekHrs<=75 AND CodeRevHrs<20
"""

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', data=df, sizes=(2, 100), alpha=0.6, edgecolor='black')

plt.title('Bubble Plot of Work week Hours vs Code Rev Hrs by Age')
plt.xlabel('Work Week Hrs')
plt.ylabel('Code Rev hours')


plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear,COUNT(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count
"""

df = pd.read_sql_query(QUERY,conn)
df=df.sort_values(by='count', ascending=False).head(5)
plt.figure(figsize=(8, 8))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)

plt.title('Pie Chart from DataFrame with Percentages')
plt.show()
```

```python
# your code goes here
QUERY = """
SELECT DatabaseWorkedWith,Respondent
FROM DatabaseWorkedWith
"""

df = pd.read_sql_query(QUERY,conn)

# Group by Respondent and count the number of unique databases each respondent worked with
respondent_database_count = df.groupby('Respondent')['DatabaseWorkedWith'].nunique().reset_index()

# Filter the respondents who worked with only one database
respondents_one_database = respondent_database_count[respondent_database_count['DatabaseWorkedWith'] == 1]

# Merge back with the original dataframe to see the corresponding databases they worked with
result = pd.merge(respondents_one_database, df, on='Respondent', how='left')

# Filter the result to return only those who worked with MySQL
mysql_only_respondents = result[result['DatabaseWorkedWith_y'] == 'MySQL']

# Count the number of respondents who worked with only MySQL
mysql_only_count = mysql_only_respondents['Respondent'].nunique()

print(mysql_only_count)
```

```python
# your code goes here
QUERY = """
SELECT LanguageDesireNextYear,COUNT(*) as count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count
"""

df = pd.read_sql_query(QUERY,conn)
df=df.sort_values(by='count', ascending=False).head(5)
print(df)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
WHERE Age>=30 AND Age<=35
group by Age
"""

df = pd.read_sql_query(QUERY,conn)
grouped_data = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True)

plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp,Age
FROM master
WHERE Age>=25 AND Age<=30
group by Age
"""

df = pd.read_sql_query(QUERY,conn)
grouped_data = df.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data, marker='o', color='blue', linestyle='-', markersize=8)

plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')

plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT DevType, Respondent
FROM DevType
"""

df = pd.read_sql_query(QUERY,conn)
print(df.head(5))
branch_counts = df['DevType'].value_counts()

branch_counts.plot(kind='barh', figsize=(8,6), color='skyblue')

plt.title('Distribution of Main Branch Categories')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.show()
```
